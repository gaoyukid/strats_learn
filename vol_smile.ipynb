```python
import numpy as np
import pandas as pd
import yfinance as yf
from scipy.stats import norm
import matplotlib.pyplot as plt
```

```python
ticker = yf.Ticker('TSLA')
options = ticker.options
chain = ticker.option_chain( options[4] )

def filter_open_interest(chain, threshold=500):
    return chain[chain['openInterest'] > threshold]


calls = filter_open_interest(chain.calls, 50)
puts = filter_open_interest(chain.puts)
```

```python
plt.figure(figsize=(12, 6))

plt.plot(puts['strike'], puts['impliedVolatility'])
```
